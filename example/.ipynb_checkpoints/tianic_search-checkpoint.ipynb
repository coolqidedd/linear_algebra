# 泰坦尼克号—生还几率与客舱等级的关系调查分析


## 不同舱等对生还率是否影响？

```python
# -*- coding: utf-8 -*- 

# Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline


# 读取原始数据    
original_data = pd.read_csv("titanic-data.csv");
original_data.info()
```

港口 |

## 生还人数与比例

```python
#总人数
total_num = len(original_data);
#总生还人数
total_survived_num = original_data['Survived'].sum()
#总未生还人数
total_no_survived_num = total_num - total_survived_num

# 图表
plt.figure(figsize = (10,5))
plt.subplot(121)
# 柱状图
sns.countplot(x='Survived', data=original_data)
#table
table_vals = [[total_num,total_survived_num,total_no_survived_num]]
col_labels = ['total','total survived','total no survived']
row_labels = ['num']
table=plt.table(cellText=table_vals,
      rowLabels=row_labels,
      colLabels=col_labels,colWidths=[0.1]*3,
      loc='bottom',bbox=[0, -0.4, 1, 0.15])
table.set_fontsize(24)
table.scale(10, 10)

plt.title('Survived Count')

plt.subplot(122)
explode = (0,0.07)
plt.pie([total_no_survived_num, total_survived_num],labels=['No Survived','Survived'],explode= explode,autopct='%1.1f%%')
plt.title('Survived Rate') 

plt.show()
```

## 不同级别客舱的人数和比例

```python
# 不同级别客舱的人数和比例

# 图表
plt.figure(figsize= (10,5))
plt.subplot(121)
#柱状图
sns.countplot(x='Pclass', data=original_data)
plt.title('Pclass Count') 

plt.subplot(122)
explode = (0.1,0,0)
# 不同级别客舱的人数和比例
survived_pclass_rate = original_data[['Pclass','Survived']].groupby(['Pclass']).count();
plt.pie(survived_pclass_rate,explode=explode,
        labels=['Pclass-1','Pclass-2','Pclass-3'],autopct='%1.1f%%')
plt.title('The survival rate in the Pclass population') 

# plt.subplot(133)
#plt.pie(original_data[['Pclass','Survived']].groupby(['Pclass']).count(),\
#        labels=['Pclass-1','Pclass-2','Pclass-3'])

plt.show()
```

## 不同级别客舱生还人数和占总生还人数的比例

```python
survived_data = original_data[original_data[ 'Survived'] == 1 ]
sum_survived_data = survived_data[['Pclass','Survived']].groupby(['Pclass']).sum()

plt.figure(figsize= (10, 5))

plt.subplot(121)
sns.countplot(x='Pclass', data=survived_data)
plt.title('Pclass Survived count') 
plt.ylabel('Survived Count')

plt.subplot(122)
explode=[0.1,0,0]
plt.pie(sum_survived_data,
        labels=['Pclass-1','Pclass-2','Pclass-3'],explode=explode,autopct='%1.1f%%')
plt.title('pclass survival rate in the survival population');
plt.show()
```

### 细节：不同舱等对应的生还、未生还人数及舱等的生还率

```python
# 不同舱等对应的生还和未生还人数及生还率
pclass1 = original_data[original_data['Pclass'] == 1]
pclass2 = original_data[original_data['Pclass'] == 2]
pclass3 = original_data[original_data['Pclass'] == 3]

plt.figure(figsize=(10,17))
plt.subplot(321)
sns.countplot(x = 'Survived', data = pclass1)
plt.title('Pclass 1 "No Survived" and "Survived"') 

plt.subplot(322)
plt.pie([pclass1['Survived'][pclass1['Survived'] == 0].count(),
         pclass1['Survived'][pclass1['Survived'] == 1].count()],
        labels=['No Survived', 'Survived'],autopct='%1.1f%%')
plt.title('Pclass 1 "No Survived" and "Survived" rate')

plt.subplot(323)
sns.countplot(x = 'Survived', data = pclass2)
plt.title('Pclass 2 "No Survived" and "Survived"') 

plt.subplot(324)
plt.pie([pclass2['Survived'][pclass2['Survived'] == 0].count(),
         pclass2['Survived'][pclass2['Survived'] == 1].count()],
        labels=['No Survived', 'Survived'],autopct='%1.1f%%')
plt.title('Pclass 2 "No Survived" and "Survived" rate')

plt.subplot(325)
sns.countplot(x = 'Survived', data = pclass3)
plt.title('Pclass 3 "No Survived" and "Survived"') 

plt.subplot(326)
plt.pie([pclass3['Survived'][pclass3['Survived'] == 0].count(),
         pclass3['Survived'][pclass3['Survived'] == 1].count()],
        labels=['No Survived', 'Survived'],autopct='%1.1f%%')
plt.title('Pclass 3 "No Survived" and "Survived" rate')

plt.show()
```

## 不同舱等对应的生还平均人数

```python
plt.figure(figsize=(5,5))
plt.subplot(111)
survived_by_pclass = original_data.groupby('Pclass')['Survived'].mean()
survived_by_pclass.plot(kind = 'bar')

plt.show()
```

# 上述图形分析得知，头等舱的用户生还几率要高于其他低等舱
